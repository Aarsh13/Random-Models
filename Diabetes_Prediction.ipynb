```python
label_counts=diabetes_database['Outcome'].value_counts()
label_counts
```

```python
label_0_data=diabetes_database[diabetes_database['Outcome']==0]
label_1_data=diabetes_database[diabetes_database['Outcome']==1]

label_0_data = label_0_data.sample(n=label_counts[1])
```

```python
diabetes_database_balanced = pd.concat([label_0_data, label_1_data])
diabetes_database_balanced.head()
```

```python
X=diabetes_database_balanced.drop(columns='Outcome',axis=1)
Y=diabetes_database_balanced['Outcome']
X
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model= svm.SVC(kernel='linear')
```

```python
model.fit(X_train, Y_train)
```
